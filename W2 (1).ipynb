```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("imdb_top_1000.csv")

runtime_mean = df["Runtime (Minutes)"].mean()
runtime_std = df["Runtime (Minutes)"].std()

threshold = runtime_mean + 2 * runtime_std

long_runtime_outliers = df[df["Runtime (Minutes)"] > threshold][["Title", "Genre", "Runtime (Minutes)"]]

print("Mean runtime:", runtime_mean)
print("Standard deviation:", runtime_std)
print("Outlier threshold:", threshold)
print(long_runtime_outliers)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("imdb_top_1000.csv")

most_voted_per_genre = df.loc[df.groupby("Genre")["Votes"].idxmax(), ["Genre", "Title", "Votes"]]

print(most_voted_per_genre.sort_values("Votes", ascending=False))

highest_voted_movie = most_voted_per_genre.sort_values("Votes", ascending=False).iloc[0]
print("Genre with the highest voted movie:", highest_voted_movie["Genre"])
print("Movie:", highest_voted_movie["Title"])
print("Votes:", highest_voted_movie["Votes"])
```
